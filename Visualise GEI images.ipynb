# Gait Energy Images - Visualisation

Gait energy images (GEI) are derived from a sequence of human silhouette frames compressed into a single binary image. This workbook provides some examples of sequences from 1) TUM dataset 2) Sequences captured at Bupa.

```python
import os
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

## TUM Dataset
The [TUM ITKP](https://www.mmk.ei.tum.de/en/misc/tum-iitkgp-gait-database/) dataset contains silhouette video sequences of 35 people walking in various scenarios. GEIs are calculated for each sequence and visualised below.

```python
GEI_DIR = 'data/gei/Normal Walk'

def get_id_and_num(filename):
    els = filename.split('_')
    id = els[1]
    num=els[3].split('.')[0]
    return id, num

files = os.listdir(GEI_DIR)
ids_and_nums = [get_id_and_num(f) for f in files]
ids = set([el[0] for el in ids_and_nums])

plt.figure(figsize= [14,100])
ids = sorted(ids)
for id in ids:
    numeric_id = int(id[2:])
    nums = sorted([el[1] for el in ids_and_nums if el[0] == id])
    for n in nums:
        ax = plt.subplot(35,4,int(n) + 4 * (numeric_id-1) )
        img = cv2.imread('{}/normal_{}_1.avi_{}.png'.format(GEI_DIR, id, n))
        plt.imshow(img)
        # Hide grid lines
        ax.grid(False)

        # Hide axes ticks
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_xlabel(id)
    


```

A number of video sequences were recorded at Bupa showing someone walking normally and walking normally whilst experiencing pain. GEIs are displayed for the 2 scenarios below.

```python
root = 'examples'
file_names = ['manu_kumar_pain_1.png', 'manu_kumar_pain_2.png', 'manu_kumar_pain_3.png', 
              'manu_kumar_nopain_1.png', 'manu_kumar_nopain_2.png', 'manu_kumar_nopain_3.png']
plt.figure(figsize= [12,10])
plt.title('Walk with no pain v Walk with pain')
for ix, fname in enumerate(sorted(file_names)):
    ax = plt.subplot(2,3, ix+1 )
    img = cv2.imread('{}/{}'.format(root, fname))
    id = fname.split('_')[2]
    plt.imshow(img)
    # Hide grid lines
    ax.grid(False)

    # Hide axes ticks
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlabel(id)
    
```
